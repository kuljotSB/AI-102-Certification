## Testing Azure Language Service QnA Solution with API Calls

### Setting up the Environment

```python
import os
import json
from dotenv import load_dotenv
import requests

load_dotenv()
prediction_url = os.getenv("PREDICTION_URL")
api_key = os.getenv("API_KEY")
```

### Starting the Chat Loop
Questions = [
    "What is Microsoft Q&A?",
    "How can i contact support?"
]

```python
while True:
    question = input("Enter your question (or type 'exit' to quit): ")
    if question.lower() == 'exit':
        break

    headers = {
        "Ocp-Apim-Subscription-Key": api_key,
        "Content-Type": "application/json"
    }

    payload = {
        "question": question
    }
    response = requests.post(prediction_url, headers=headers, json=payload)
    answer_data = response.json()
    print("Answer:", answer_data['answers'][0]['answer'])
```
